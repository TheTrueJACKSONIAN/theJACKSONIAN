```
import pandas as pd
import numpy as np
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```
raw_data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
raw_data
```

```
raw_data.columns
```

```
column_dict = {'date':'date', 'Appliances':'Appliances', 'lights':'lights', 'T1':'kitchen_temp', 'RH_1':'kitchen_humidity', 'T2':'livingRm_temp', 'RH_2':'livingRm_humidity',
               'T3':'laundryRm_temp', 'RH_3':'laundryRm_humidity', 'T4':'officeRm_temp', 'RH_4':'officeRm_humidity', 'T5':'bathroom_temp', 'RH_5':'bathroom_humidity',
               'T6':'outside_temp(ns)', 'RH_6':'outside_humidity(ns)', 'T7':'ironingRm_temp', 'RH_7':'ironingRm_humidity',
               'T8':'teenRm2_temp', 'RH_8':'teenRm2_humidity', 'T9':'parentsRm_temp', 'RH_9':'parentsRm_humidity', 'T_out':'outside_temp(chievres)', 'Press_mm_hg':'outsidePres(chievres)_mmHhg', 'RH_out':'outside_humidity(chievres)',
               'Windspeed':'windspeed(chievres)', 'Visibility':'visibility(chievres)', 'Tdewpoint':'Tdewpt(chievres)', 'rv1':'randvar1', 'rv2':'randvar2'}

data_col_renamed = raw_data.copy()
data_col_renamed = data_col_renamed.rename(columns = column_dict)
data_col_renamed
```

```
data_col_renamed.describe()
```

```
data_no_date_lights = data_col_renamed.copy()
data_no_date_lights = data_no_date_lights.drop(columns = ['date','lights'])
data_no_date_lights
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_normalized = pd.DataFrame(scaler.fit_transform(data_no_date_lights), columns = data_no_date_lights.columns)
data_normalized
```

```
data_features = data_normalized.drop(columns = ['Appliances'])
data_features
```

```
simple_linear_reg_df = data_features[['livingRm_temp', 'outside_temp(ns)']]
simple_linear_reg_df
```

```
ar1 = np.array(data_features['livingRm_temp']).reshape(-1,1)
ar2 = np.array(data_features['outside_temp(ns)']).reshape(-1,1)

x_trainer, x_tester, y_trainer, y_tester = train_test_split(ar1, ar2, test_size = 0.3, random_state = 42)

linReg = LinearRegression()
linReg.fit(ar1, ar2)
```

```
data_features['outside_temp(ns)']
```

```
data_target = data_normalized['Appliances']
data_target
```

```
x_train, x_test, y_train, y_test = train_test_split(data_features, data_target, test_size = 0.3, random_state = 42)
```

```
linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)
```

```
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(x_train, y_train)
```

```
lasso_reg = Lasso(alpha = 0.001, max_iter = 2000)
lasso_reg.fit(x_train, y_train)
```

```
elnet_reg = ElasticNet(alpha = 0.00001, l1_ratio = 1.0)
elnet_reg.fit(x_train, y_train)
```

```
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
```

```
linear_reg_weights = get_weights_df(linear_reg, x_train, 'Linear_Weights')
ridge_reg_weights = get_weights_df(ridge_reg, x_train, 'Ridge_Weights')
lasso_reg_weights = get_weights_df(lasso_reg, x_train, 'Lasso_Weights')
elasticNet_reg_weights = get_weights_df(elnet_reg, x_train, 'ElasticNet_Weights')
```

```
weights_data0 = pd.merge(linear_reg_weights, ridge_reg_weights, on='Features')
weights_data1 = pd.merge(weights_data0, lasso_reg_weights, on = 'Features')
weights_data = pd.merge(weights_data1,  elasticNet_reg_weights, on = 'Features')
weights_data
```

```
linear_prediction = linear_reg.predict(x_test)
ridge_prediction = ridge_reg.predict(x_test)
lasso_prediction = lasso_reg.predict(x_test)
elnet_prediction = elnet_reg.predict(x_test)
```

```
linReg_Quiz = linReg.predict(x_tester)
```
